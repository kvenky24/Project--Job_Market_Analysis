### Data Cleaning

**Checking Missing Values**

```python
df.isnull().sum()
```

```python
df=df.fillna(df.mode().iloc[0])
```

```python
df.isnull().sum()
```

**Removing Duplicate Values**

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```
